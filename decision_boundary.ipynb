```python
import numpy as np
import matplotlib.pyplot as plt

def disc(x, u, v):
    return (np.exp(np.dot(x, u)) / (np.exp(np.dot(x, u)) + 3*np.exp(np.dot(x, v)))) > 0.5
```

```python
x = np.linspace(-10, 10, 100); y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
inp = np.array([X.ravel(), Y.ravel()]).T

u = np.array([0.2, -0.4]); v = np.array([0.2, 0.5])

# np.exp(np.dot(inp, u)).shape
oc = disc(inp, u, v)
plt.scatter(inp[:, 0], inp[:, 1], c=oc, s=1)
plt.show()
```
