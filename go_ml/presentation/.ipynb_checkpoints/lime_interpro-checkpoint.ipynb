```python
import torch
import numpy as np
import json
import matplotlib.pyplot as plt
import pandas as pd
import pickle
```

```python
train_path = "/home/andrew/cafa5_team/data/"
with open(f"{train_path}/cafa_dataset/go_terms.json", "r") as f:
    go_terms = json.load(f)
domain_df = pd.read_csv('../../data/superfamily_matches.tsv', sep='\t')
domain_df = domain_df[~domain_df['GO Terms'].isna()]
domain_terms = [gt.split(',') for gt in domain_df['GO Terms']]
domain_term_ind = [[go_terms.index(t) for t in r if t in go_terms] for r in domain_terms]
domain_df['GO Ind'] = domain_term_ind
domain_df = domain_df[domain_df['Length'] < 800]

from go_bench.load_tools import load_protein_sequences
prot_seq, prot_id = load_protein_sequences('../../data/interpro_superfamilies.fasta', prot_whitelist=set(domain_df['Accession Num']))

annotated_indices = []

for domain in domain_df['Protein Locations']:
    domain_indices = []
    ranges = domain.split(",")
    for start_end in ranges:
        start, end = map(int, start_end.split(".."))
        indices = np.arange(int(start), int(end) + 1)
        domain_indices.extend(indices)
    annotated_indices.append(domain_indices)
    

domain_df['AnnotatedIndices'] = annotated_indices


from Bio import SeqIO

def create_sequence_dict(fasta_file):
    sequences = {}
    
    for record in SeqIO.parse(fasta_file, "fasta"):
        accession_code = record.id.split("|")[1]
        sequences[accession_code] = str(record.seq)
    
    return sequences

fasta_file_path = '../../data/interpro_superfamilies.fasta'

sequences_dict = create_sequence_dict(fasta_file_path)
domain_df['seq'] = domain_df['Accession Num'].str.strip().map(sequences_dict)

domain_df = domain_df[~domain_df['GO Ind'].isna()]
domain_df = domain_df[domain_df['GO Ind'].apply(lambda x: len(x) > 0)]

import transformers
tokenizer = transformers.AutoTokenizer.from_pretrained('facebook/esm2_t6_8M_UR50D')

# mulitple domains per protein 

def domain_iterator():
     for i, name, source_db, type, integrated, member_DB, goterms, acc, len, protein_loc, pid, sf_acc, go_ind, annot_ind, seq in domain_df.itertuples():
          inputs = tokenizer.batch_encode_plus([seq], add_special_tokens=True, padding='max_length',
                                             truncation=True, return_attention_mask=True, max_length=800)
          yield {'prot_id': pid, 'annot_ind': annot_ind, 'go_ind': go_ind, 'seq': seq, 'seq_ind': inputs['input_ids'], 'mask': inputs['attention_mask']}
seq_len = torch.LongTensor([len(r['seq']) for r in domain_iterator()])
```

```python
from go_ml.aa_params import aromatic_scale, pKs_scale, gravy_scale, protein_weight_scale
from sklearn.preprocessing import StandardScaler
import numpy as np

def residue_emb(sequence):
    sequence_embeddings_list = []
    for s in sequence:
        sequence_embeddings_list.append([aromatic_scale[s], pKs_scale[s], gravy_scale[s], protein_weight_scale[s]])
    sequence_embeddings = np.array(sequence_embeddings_list)
    return sequence_embeddings

seq_emb_l = []
for domain in domain_iterator():
    seq = domain['seq']
    seq_emb = residue_emb(seq)
    seq_emb_l.append(seq_emb)
res_emb = np.concatenate(seq_emb_l, axis=0)
residue_scaler = StandardScaler().fit(res_emb)

res_emb.shape
```

```python
import goatools
from goatools.obo_parser import GODag
godag = GODag('../go-basic.obo')
from goatools.godag.go_tasks import get_go2parents
optional_relationships = set()
go2parents_isa = get_go2parents(godag, optional_relationships)
from goatools.gosubdag.gosubdag import GoSubDag

gosubdag_r0 = GoSubDag(['GO:0018683'], godag, prt=None)
go_ancestors = gosubdag_r0.rcntobj.go2ancestors['GO:0018683']
go_ancestors.add('GO:0018683')
```

```python
def get_parent_ind(go_ind):
    go_term = go_terms[go_ind]
    if(not go_term in go2parents_isa):
        return go_ind    
    parents = go2parents_isa[go_term]
    if(len(parents) == 0):
        return go_ind
    parent_inds = [term_ind_map[p] for p in parents if p in term_ind_map]
    if(len(parent_inds) == 0):
        return go_ind
    return parent_inds[0]
```

```python
def get_ancestor_ind(go_ind):
    go_term = go_terms[go_ind]
    if(not go_term in godag):
        return []
    gosubdag_r0 = GoSubDag([go_term], godag, prt=None)
    go_ancestors = gosubdag_r0.rcntobj.go2ancestors[go_term]
    go_ancestors.add(go_term)
    ancestor_ind = [term_ind_map[anc] for anc in go_ancestors if anc in term_ind_map]
    return ancestor_ind
```

```python
mut_residue_emb_l = []
for i, r in enumerate(domain_iterator()):
    prot_mut_l = lime_dict[r["prot_id"]][0] # mutated sequences 
    mut_residue_emb = []
    for mut in prot_mut_l:
        pm = mut[0]
        mut_residue_emb.append(residue_scaler.transform(residue_emb(pm)))
    mut_residue_emb = np.array(mut_residue_emb)
    mut_residue_emb_l.append(mut_residue_emb)
    if i % 10 == 0:
        print(i)
```

```python
from sklearn.linear_model import LinearRegression
from scipy.sparse import csr_array, coo_array

lime_models = []
for i, r in enumerate(domain_iterator()):
    go_ind = r['go_ind'][0]
    mut_residue_emb = mut_residue_emb_l[i]
    coef = np.zeros(mut_residue_emb[0].shape)
    local_pred = lime_dict[r["prot_id"]][1]
    lime_coo = coo_array(local_pred)
    nz_r, nz_c = lime_coo.coords
    lime_eval = lime_coo.toarray()
    zero_update_mask = np.zeros_like(lime_eval) - 9.0
    zero_update_mask[nz_r, nz_c] = 0
    lime_eval += zero_update_mask
    local_pred = lime_eval[:,go_ind]
    
    if(local_pred.sum()==0):
        go_ind = get_parent_ind(go_ind)
        local_pred = lime_eval[:,go_ind]

    lime_model = LinearRegression()
    # print(local_pred.shape)
    flat_mut_residue_emb = mut_residue_emb.reshape(1500, -1)
    # print(flat_mut_residue_emb.shape)
    lime_model.fit(flat_mut_residue_emb, local_pred)
    # print(lime_model.coef_.shape)
    coef += lime_model.coef_.reshape(mut_residue_emb[0].shape)
    lime_models.append(coef)

    if i % 10 == 0:
        print(i)
```

```python
def precision_top_k(true_indices, coef, k):
    top_k_predicted_indices = np.argsort(coef)[-k:]
    true_positive_indices = set(true_indices)
    true_positives = len([idx for idx in top_k_predicted_indices if idx in true_positive_indices])
    precision = true_positives / k
    return precision


def get_avg_precision(iterator, lime_models, aggregate_type, k):
    precisions = np.zeros(len(lime_models))
    for i, r in enumerate(iterator()):
        coef = lime_models[i]
        if aggregate_type == "mag":
            coef = np.sum(coef, axis=1)
        elif aggregate_type == "abs":
            coef = np.sum(np.abs(coef), axis=1)
        else:
            print("wrong type")
        annot_ind = r["annot_ind"] 
        annot_ind = [i - 1 for i in annot_ind]
        precisions[i] = precision_top_k(annot_ind, coef, k)
    return np.mean(precisions)
        
```

```python
import torch
import numpy as np

def mutate(seq, lambda_=15):
    amino_acids = "ACDEFGHIKLMNPQRSTVWY"
    mut_seq = np.array(list(seq))
    num_changes = np.random.poisson(lambda_)
    indices_to_change = np.random.choice(len(seq), num_changes, replace=False)
    mut_seq[indices_to_change] = np.random.choice(list(amino_acids), num_changes)
    mut_seq = ''.join(mut_seq)
    return mut_seq, num_changes

def mutate_ind(seq, key_ind, lambda_=15):
    amino_acids = "ACDEFGHIKLMNPQRSTVWY"
    mut_seq = np.array(list(seq))
    num_changes = np.random.poisson(lambda_)
    indices_to_change = np.random.choice(len(key_ind), num_changes, replace=False)
    mut_seq[key_ind[indices_to_change]] = np.random.choice(list(amino_acids), num_changes)
    mut_seq = ''.join(mut_seq)
    return mut_seq, num_changes

def mut_batch(prot_id, seq, key_ind, lambda_=15, batch_size=100):
    mut_seq = [mutate_ind(seq, key_ind, lambda_=lambda_)[0] for _ in range(batch_size)]
    inputs = tokenizer.batch_encode_plus(mut_seq, add_special_tokens=True, padding='max_length',
                    truncation=True, return_attention_mask=True, max_length=1024) 
    return {'seq': mut_seq, 'prot_id': [prot_id]*batch_size, 'seq_ind': inputs['input_ids'], 'mask': inputs['attention_mask']}
```

```python
from scipy.sparse import csr_array, vstack
def get_evals(model, mut_seq_dataset, batch_size=40):
    d = mut_seq_dataset
    device = model.device
    eval_l = []
    with torch.no_grad():
        for i in range(0, len(d['seq_ind']), batch_size):
            seq_ind, mask =  torch.tensor(d['seq_ind'][i:i+batch_size]).to(device), torch.BoolTensor(d['mask'][i:i+batch_size]).to(device)
            logit_preds = model.forward(seq_ind, mask)
            logit_preds = logit_preds * (logit_preds > -9)
            sparse_preds = csr_array(logit_preds.cpu().numpy())
            eval_l.append(sparse_preds)
            if(i % (2*batch_size) == 0 and i > 0):
                print(f"{100*i/len(d['seq_ind'])}% Eval")
                break
    eval_results = vstack(eval_l)
    return eval_results
model_evals = get_evals(model, mut_seq_dataset, batch_size=40)
```
