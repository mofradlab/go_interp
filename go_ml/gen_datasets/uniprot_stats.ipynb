```python
from Bio import SeqIO
with open('/home/andrew/GO_interp/data/uniprotkb_2024_12_16.fasta', 'r') as f:
    sequences = [str(s.seq) for s in SeqIO.parse(f, 'fasta')]
from collections import Counter
aa_counts = Counter(''.join(sequences))
```

```python
tc = sum(x for x in aa_counts.values())
aa_freq = {aa:c/tc for aa, c in aa_counts.items()}
```

```python
import json
with open('/home/andrew/GO_interp/data/uniprot_aa_freq.json', 'w') as f:
    json.dump(aa_freq, f)
```
