```python
import torch
test_seq = torch.zeros((50,))

def gen_mut_batch(seq, num_samples, max_muts):
    L = seq.shape[0]
    mut_sites = torch.randint(0, L, (num_samples*max_muts,))
    mut_batch_ind = torch.tile(torch.arange(0, num_samples).reshape(-1, 1), (1, max_muts)).flatten()
    base_residues = seq[mut_sites]
    mut_mask = torch.rand(mut_sites.shape) > 0.2
    mut_residues = torch.randint(0, 23, mut_sites.shape)*mut_mask + base_residues*(~mut_mask)
    mut_batch = torch.tile(seq.reshape(1, -1), (num_samples, 1))
    mut_batch[mut_batch_ind, mut_sites] = mut_residues
    return mut_batch

mut_batch = gen_mut_batch(test_seq, 12, 5)
print(mut_batch)
```
